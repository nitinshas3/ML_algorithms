```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_dataset.csv')
print(df.head())
```

```python
def costFunction(m,b,df):
    error = 0
    for i in range(df.shape[0]):
        error+=(((m*df.loc[i,'YearsExperience']) + b ) - df.loc[i,'Salary'])**2
    error/=df.shape[0]
    return error
```

```python
def gradientdDescent(m_now,b_now,learning_rate,df):
    m_gradient = b_gradient = 0
    for i in range(df.shape[0]):
        x = df.loc[i, 'YearsExperience']
        y = df.loc[i, 'Salary']
        m_gradient+=x*(y-(m_now*x + b_now))
        b_gradient += (y - (m_now*x +b_now))
    m_gradient*= (-2/df.shape[0])
    b_gradient*= (-2/df.shape[0])
    m_new = m_now - m_gradient*learning_rate
    b_new = b_now - b_gradient*learning_rate
    return m_new,b_new
```

```python
m = 0
b = 0
iterations = 1000
learning_rate = 0.001
```

```python
for i in range (iterations):
    m,b = gradientdDescent(m,b,learning_rate,df)
```

```python
print(m,b)
```

```python
x = df['YearsExperience']
y = df['Salary']
y_pred = m * x + b

plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', label='Predicted')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
