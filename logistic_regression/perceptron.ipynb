```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('placement-dataset.csv')
print(df.head())
```

```python
placed = df[df['placement'] == 1]
not_placed = df[df['placement'] == 0]


plt.scatter(placed['cgpa'], placed['iq'], color='blue', label='Placed')
plt.scatter(not_placed['cgpa'], not_placed['iq'], color='red', label='Not Placed')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Placement Status')
plt.legend()
plt.show()
```

```python
x =np.array( df.iloc[:,1:3])
x=np.insert(x,0,1,axis=1)
print(x[:5])
y = df.iloc[:,-1:]
y = np.array(y)
print(y[:5])
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
def perceptron(x,y):
    
    
    lr = 0.1
    weights = np.ones(x.shape[1])
    for i in range (1000):
        j = np.random.randint(0,x.shape[0])
        y_pred = step(np.dot(x[j],weights))
        weights = weights + lr*(y[j] - y_pred)*x[j]
    return weights  
```

```python
print(x)
```

```python
weightsfinal = perceptron(x,y)
print(weightsfinal)
m = -1*(weightsfinal[1]/weightsfinal[2])
b = -1*(weightsfinal[0]/weightsfinal[2])
```

```python
x_vals = np.linspace(df['cgpa'].min(), df['cgpa'].max(), 100)
y_vals = m * x_vals + b
plt.scatter(placed['cgpa'], placed['iq'], color='blue', label='Placed')
plt.scatter(not_placed['cgpa'], not_placed['iq'], color='red', label='Not Placed')
plt.xlabel('CGPA')
plt.ylabel('IQ')
plt.title('Placement Status')

# Plot the decision boundary
plt.plot(x_vals, y_vals, color='green', label='Decision Boundary')

# Show legend and plot
plt.legend()
plt.grid(True)
plt.show()
```
